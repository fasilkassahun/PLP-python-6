## Exploring the structure

```python

print("Shape:", df.shape)
print("\nInfo:")
print(df.info())
print("\nDescription:")
print(df.describe())
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Option 1: Drop missing values
df_clean = df.dropna()

# Option 2: Fill missing values (example: fill numeric with mean)
df_filled = df.fillna(df.mean(numeric_only=True))
```

#  Task 3

## importing

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Line Chart (trend over time)


```python
plt.figure(figsize=(8,5))
plt.plot(df.index, df["sepal_length"], label="Sepal Length")
plt.title("Line Chart - Sepal Length over Samples")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.show()
```

##  Bar Chart (average petal length per species)

```python
plt.figure(figsize=(8,5))
sns.barplot(x="species", y="petal_length", data=df, estimator="mean")
plt.title("Average Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Length (cm)")
plt.show()
```

## Histogram (distribution of sepal length)

```python
plt.figure(figsize=(8,5))
plt.hist(df["sepal_length"], bins=20, color="skyblue", edgecolor="black")
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

##  Scatter Plot (relationship between two features)

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=df)
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.show()
```
